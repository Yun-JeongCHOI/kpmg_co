```python
# label_remove에 있는 파일리스트 불러오기   
import os
import pandas as pd
import numpy as np
import re
import glob
import json
import pickle
import time
import datetime

# path는 label_remove에 있는 파일리스트 불러오기
path = 'D:/jyp/kpmg_co/1. preprocessing/1-1. labeling_preprocessing/output/train/label_remove'
```

```python
# 파일리스트 불러오기
file_list = os.listdir(path)
```

```python
file_list = [file for file in file_list if file.endswith(".csv")]
```

```python
for i in file_list:
    df = pd.read_csv(path + '/' + i)
    df = df.drop(['Unnamed: 0'], axis=1)
    # df의 label column의 값이 4인 경우 1로 바꾸기
    df.loc[df['label'] == 4, 'label'] = 1
    # df의 label column의 값이 5인 경우 2로 바꾸기
    df.loc[df['label'] == 5, 'label'] = 2
    # df의 label column의 값이 6인 경우 3로 바꾸기
    df.loc[df['label'] == 6, 'label'] = 3
    
    # df의 label column의 값이 9일 경우 0으로 변환
    df.loc[df['label'] == 9, 'label'] = 0
    
    # df의 label column의 값이 20일 경우 2으로 변환
    df.loc[df['label'] == 20, 'label'] = 2
    # df의 index 초기화
    df = df.reset_index(drop=True)
    
    # df 저장
    
    # output 폴더가 없으면 생성
    if not os.path.exists('D:/jyp/kpmg_co/1. preprocessing/1-1. labeling_preprocessing/output/train/label_remove/output/v3'):
        os.makedirs('D:/jyp/kpmg_co/1. preprocessing/1-1. labeling_preprocessing/output/train/label_remove/output/v3')
    df.to_csv('D:/jyp/kpmg_co/1. preprocessing/1-1. labeling_preprocessing/output/train/label_remove/output/v3' + i, encoding='utf-8-sig')
```

```python
file_path = 'D:/jyp/kpmg_co/1. preprocessing/1-1. labeling_preprocessing/output/train/label_remove/output/'
```

```python
# file path에 있는 파일 리스트 불러오기 
file_list = os.listdir(file_path)
file_list
```

```python
# file_list에 있는 파일들을 하나씩 불러와서 하나의 df로 만들기

corpe_df = pd.DataFrame()
for i in file_list:
    df = pd.read_csv(file_path + i)
    df = df.drop(['Unnamed: 0'], axis=1)
    df = df.reset_index(drop=True)
    # corpe_df에 추가하기
    corpe_df = pd.concat([corpe_df, df], axis=0)
```

```python
# date 기준 정렬    
corpe_df = corpe_df.sort_values(by=['date'], axis=0)
```
