```python
# 전처리를 하기 위한 라이브러리를 불러옵니다.

import pandas as pd
import os
```

### 전처리

```python
# 보내드린 raw 파일을 불러옵니다.

path = "D:/jyp/naver_news_2/esg/sin/"
pre_df = pd.read_csv(path + 'sin_raw.csv')
```

```python
# csv 파일을 불러왔을 때, unnamed: 0 이라는 불필요한 열이 생기는데, 이를 제거해줍니다.

pre_df = pre_df.drop(['Unnamed: 0'], axis=1)
```

```python
# 아래 셀을 한 번 돌릴 때마다 해당 셀을 실행해주세요.

pre_df = pre_df.reset_index(drop=True)
pre_df
```

```python
###############################################################################################################
# 전처리할 키워드를 넣는 곳입니다. if 옆에 있는 '포토'를 원하는 키워드로 바꿔주세요.
# 한번에 여러 키워드를 지정하고 싶다면, or을 사용하면 됩니다.
# ex) if '포토' or '사진' in pre_df['title'][j]:
# 이런 식으로 하면, 포토와 사진이라는 키워드가 들어간 기사들이 전부 제거됩니다.
# 한번에 잘 제거가 되지 않으니 2번 정도 돌려주세요.
# 한번 돌리실 때 마다 , 위에 있는 셀을 실행해주셔서 rows 개수가 줄어든 것을 확인해주시고, 다시 해당 셀을 실행해주세요. 
###############################################################################################################

for j in range(len(pre_df['title'])):
    if '포토' in pre_df['title'][j]:
        # j 행 제거 후, 인덱스 재정렬
        pre_df = pre_df.drop(j)
        pre_df = pre_df.reset_index(drop=True)
        print(j)
```

```python
# 저장할 파일 이름을 지정해주시면 됩니다.

outputFileName = 'sin_raw.csv'
pre_df.to_csv(path+'output/'+outputFileName)
```

### 슬라이싱

```python
# 전처리된 파일을 불러옵니다.

path_pre = "D:/jyp/kpmg/1. preprocessing/"
pre_pre_df = pd.read_csv(path_pre + 'sin_output.csv')
```

```python
pre_pre_df = pre_pre_df.drop(['Unnamed: 0'], axis=1)
pre_pre_df
```

```python
# 2020-01-01부터 2023-01-15까지 business day에 대한 날짜를 생성합니다.  (2020-01-01 ~ 2023-01-15)

date = pd.date_range(start='2020-01-01', end='2023-01-15', freq='B')
date = pd.DataFrame(date)
# date를 list로 바꾸기  (2020-01-01 ~ 2023-01-15)
date_list = date[0].tolist()
date_list = [i.strftime("%Y-%m-%d") for i in date_list]
```

```python
date_list
```

```python
# 날짜별 10개씩 추가하는 코드

n_df = pd.DataFrame()

# df와 date_list의 date column이 일치할 경우 new_df에 추가  (일치하지 않을 경우 new_df에 추가하지 않음)
for i in range(len(date_list)):
    count = 0
    for j in range(len(pre_pre_df['date'])):
        if count == 10:
            pass
        elif date_list[i] == pre_pre_df['date'][j]:
            n_df = n_df.append(pre_pre_df.loc[j], ignore_index=True)
            count +=1
            print(i, j)
```

```python
# n_df의 column을 'date' 'title' 'contents' 순으로 재정렬

n_df = n_df[['date', 'title', 'contents']]
```
